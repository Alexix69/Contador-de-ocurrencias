```python
def getFile(fileName):
    file = open(fileName, encoding='utf8')
    wordList = []
    line = file.readline()
    while line != "$":
        for word in line.split():
            wordList.append(word)
        line = file.readline()
    file.close()
    return wordList    

wordlist = getFile("harry.txt")
repetitionlist = []
for i in wordlist:
    if (i not in repetitionlist):
        repetitionlist.append(i)
        print(i + ": ")
        counter = 0
        for j in wordlist:
            if (j == i):
                counter += 1
        print(str(counter) + "\n")
```
